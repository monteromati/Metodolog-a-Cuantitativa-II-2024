En general, trabajar√°s en Python usando un `DataFrame`, que significa cuadro de datos. Un `DataFrame` es un objeto que almacena un conjunto de series que, a su vez, pueden ser listas que almacenan informaci√≥n de tipo num√©rica, categ√≥rica, entre otros. Com√∫nmente, las series dentro de un `DataFrame` se conocen como columnas, las cuales almacenan atributos de la unidad de observaci√≥n de nuestro conjunto de datos representados en las filas. Por ejemplo, las filas pueden ser estudiantes, escuelas, pa√≠ses, mientras que las columnas ser√°n sus atributos (e.g., g√©nero).

Aqu√≠ un ejemplo de c√≥mo crear un `DataFrame` en Python:

```python
import pandas as pd

# Creando un DataFrame
data = {
    "Nombre": ["Ana", "Luis", "Marta"],
    "Edad": [23, 45, 35],
    "G√©nero": ["Femenino", "Masculino", "Femenino"]
}

df = pd.DataFrame(data)
print(df)
```

En este ejemplo, el `DataFrame` `df` almacena tres columnas (Nombre, Edad, G√©nero) con informaci√≥n sobre diferentes individuos.

**Ejercicio 6**

```python
import statsmodels.api as sm

y = CarTab['qsec']
X = CarTab[['hp', 'wt']]
X = sm.add_constant(X)

mod1 = sm.OLS(y, X).fit()
print(mod1.summary())
```
